```python
x_values = np.random.random(15)
y_values = np.random.random(15)
points = np.vstack([x_values, y_values]).T
fig, ax = plt.subplots()

def distance(p1, p2):
    return np.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

d_points = 0
p1 = 0
p2 = 0
for i, pi in enumerate(points):
    for j, pj in enumerate(points):
        d = distance(pi, pj)
        if d > d_points:
            d_points = d
            p1 = pi
            p2 = pj
            
plt.scatter(*points.T, color='0.4')
plt.scatter(*p1, color='b')
plt.scatter(*p2, color='b')
plt.plot([p1[0], p2[0]], [p1[1],p2[1]], color='b')
ax.set_aspect('equal')
```
